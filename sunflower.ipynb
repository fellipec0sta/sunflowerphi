```python
import numpy as np
import math
import pandas as pd
import seaborn as sns
```

```python
sns.set()
phi = (math.sqrt(5)-1)/2
n = 2000 #numero de sementes do girassol
```

```python
rho = []
theta =[]

for i in range(0, n):
    rho.append(pow(i,phi)) 
    theta.append(i*2* np.pi * phi)     
    
df = pd.DataFrame({'rho': rho, 'theta': theta})
```

```python
g = sns.FacetGrid(df, subplot_kws=dict(projection='polar'), height=5.5,sharex=False, sharey=False, despine=False)
g.map(sns.scatterplot, "theta", "rho")
```
